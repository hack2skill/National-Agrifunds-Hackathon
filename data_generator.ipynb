```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Number of data points
num_data_points = 3000

# List of unique mobile numbers
unique_mobile_numbers = [f"9{random.randint(100000000, 999999999)}" for _ in range(50)]

# Timestamp range
start_timestamp = datetime(2023, 7, 1,0,0,0)
end_timestamp = datetime(2023, 8, 31, 23, 59, 59)

# Locations
locations = ["Ladakh", "Tawang", "Kanyakumari", "Jaisalmer", "Darjeeling", "Gopalpur"]

# Generate dummy data
data = []
for _ in range(num_data_points):
    user_mobile = random.choice(unique_mobile_numbers)
    recipient_mobile = random.choice(unique_mobile_numbers)
    time_stamp = start_timestamp + timedelta(seconds=random.randint(0, int((end_timestamp - start_timestamp).total_seconds())))
    amount = random.randint(20, 5000)
    location = random.choice(locations)
    payment_status = random.choice(["Success", "Fail"])
    
    data.append([user_mobile, recipient_mobile, time_stamp, amount, location, payment_status])

# Create a DataFrame
df = pd.DataFrame(data, columns=["user_mobile", "recipient_mobile", "time_stamp", "amount", "location", "payment_status"])

```
